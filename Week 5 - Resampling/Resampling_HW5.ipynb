#### Question 1

We will derive the probability that a given observation is part of a bootstrap sample. Suppose that we obtain a bootstrap sample from a set of `n` observations. Don’t forget that samples are obtained with replacement.

  a) What is the probability that the first bootstrap observation is not the $jth$ observation from the original sample? Justify your answer.

> **_Answer:_**  Given n observations in the original sample, Probability of any observation from the original sample that will be part of bootstrap sample is 1/n.
>
> So, the probability that the first bootstrap observation is the jth observation from the original sample is 1/n
>
> Hence, the probability that the first bootstrap observation is _not_ the jth observation from the original sample is 1- (1/n) = (n-1)/n

  b) What is the probability that the second bootstrap observation is not the $jth$ observation from the original sample?

> **_Answer:_**  Since Bootstrapping happens with _Sampling with replacement_, Probabilty of an observation getting picked is going to be picked is  the same for all the iteration.
>
> As explained in a), the probability that the second bootstrap observation is not the jth observation from the original sample will also be ,  1- (1/n) = (n-1)/n

  c) Argue that the probability that the $jth$ observation is not in the bootstrap sample is $(1 − 1/n)^n$.

> **_Answer:_**  From a) & b) , we know that the probability that the jth observation is not taken in any given iteration in the bootstrap sample is 1- (1/n).
>
> P(jth observation not in bootstrap sample) = P(jth observation not in 1st iteration) * P(jth observation not in 2nd iteration) * .. * P(jth observation not in nth iteration)
>
> P(jth observation not in bootstrap sample) = (1- (1/n)) * (1- (1/n))* .. * (1- (1/n))
>
> P(jth observation not in bootstrap sample) = $(1- (1/n))^n$ 

  d) When `n = 5`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  From c), we know that P(jth observation not in bootstrap sample) = $(1- (1/n))^n$
>
> So, P(jth observation in bootstrap sample) = $ 1- (1- (1/n))^n$
>
> Substituting value for n = 5, P(jth observation in bootstrap sample) =  $1- (1- (1/5))^5$ = $1- (4/5)^5$ = 1 - 0.3276 = 0.6724
> 

  e) When `n = 100`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_**  P(jth observation in bootstrap sample) = $ 1- (1- (1/n))^n$.
>
> Substituting value for n = 100, P(jth observation in bootstrap sample) =  $1- (1- (1/100))^100$ = $1- (99/100)^100$ = 1 - 0.366 = 0.6339
> 

  f) When `n = 1000`, what is the probability that the $jth$ observation is in the bootstrap sample?

> **_Answer:_** P(jth observation in bootstrap sample) = $ 1- (1- (1/n))^n$.
>
> Substituting value for n = 1000, P(jth observation in bootstrap sample) =  $1- (1- (1/1000))^1000$ = $1- (999/1000)^1000$ = 1 - 0.367 = 0.632.

#### Question 2

The following questions are relative to k-fold cross-validation.

  a) Explain how k-fold cross-validation is implemented.

> **_Answer:_**  In k-fold cross-validation, the training dataset is split into k folds/ groups. In the first iteration, the 1st fold will be held for validation and the model is fitted on the remaining 4 folds. Then, the model is evaluated for MSE on the left out fold.
>
> This process is repeated k times, and in each iteration, different fold will be left out for validation and the model is fitted on the rest of the folds. At the end of k iteration, we will get k-MSE values.
>
> The k-fold cross validation test error rate would be the average of the MSE obtained in k iteration.
>  

  b) How would you choose the analyzed k? Does that matter? (please cite your sources)

> **_Answer:_**  There are multiple factors which needs to be considered when choosing the correct k-value in k-fold cross validation.
> 1. Computational cost
>    When the number of observations are high, and if choosing a very high k -value will increase the computional cost, as the number of times the model needs to be fitted would be very high.
> 
> 2. Bias - Variance trade off - Apart from the computational cost, we also need to consider the Bias -variance trade off in choosing the optimal k-value. Choosing a very low k-value means that the training dataset will be less and the model will be more biased. On the other hand, choosing a high k -value means that the training dataset will overlap across k iteration, which will lead to high variance in the output due to its high correlation.
>
> So, choosing an optimal k-value of k =5 or 10 will be provide balance of bias and variance.
>

  c) What are the advantages and disadvantages of k-fold cross-validation relative to *the validation set* approach?

> **_Answer:_**  Advantages of k-fold CV relative to *the validation set* :
> 1. Validation set approach splits the data into two. So, the model will be fitted on very less data, and the model might not be able to capture all the relationship. This will make the model more biased. In k-fold cross-validation, choosing a optimal k-value of 5 or 10 will provide the model with more training data, and the k-fold method will be less biased compared to validation approach
> 2. Since the training data is less in the validation approach, the validation set approach can lead to overestimates of the test error rate. The k-fold method can provide better estimate of the test error rate with optimal k-value
>
> Disadvantage of k-fold CV relative to *the validation set*
> 1. Computational cost of the validation approach is very less as the model is fitted only once. With k-fold, the model will be fitted many times, increasing the computational cost.

  d) What are the advantages and disadvantages of k-fold cross-validation relative to *LOOCV*?

> **_Answer:_**  Advantages of k-fold CV relative to *LOOCV*:
> 1. Computational cost of the LOOCV is very high as the model is fitted will be n times. When n becomes large, the computational cost becomes very high. With k-fold, the model will be fitted based on the chosen k value, choosing an optimal k -value,computational cost will be reduced.
> 2. LOOCV fits the model on n-1 observation. So, the model will be fitted on almost similar data on every iteration, thus the output will be highly correlated. Hence LOOCV approach tends to have high variance. In k-fold cross-validation, choosing a optimal k-value of 5 or 10 will reduce the correlation in the output. Hence k-fold cross-validation approach will have less variance compared to LOOCV.
>
> Disadvantage of k-fold CV relative to *LOOCV*
> 1. LOOCV approach will have less bias compared to k-fold cross-validation, as the LOOCV approach uses almost all the observation while training the model. Hence LOOCV will be almost unbiased compared to k-fold CV method.

```python

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})

# Number of folds
n_folds = 20

# Polynomial degrees
poly_degrees = range(1, 11)

# KFold cross-validation
kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
mse_estimates = []

# Loop over polynomial degrees
for q in poly_degrees:
    y_hat = np.zeros(n)
    
    for train_idx, test_idx in kf.split(data):
        # Fit on K-1 folds
        poly = PolynomialFeatures(degree=q)
        X_train = poly.fit_transform(data.iloc[train_idx][['x']])
        X_test = poly.transform(data.iloc[test_idx][['x']])
        
        ols = sm.OLS(data.iloc[train_idx]['y'], X_train).fit()
        
        # Predict on the Kth fold
        y_hat[test_idx] = ols.predict(X_test)
    
    # Get the MSE estimate
    mse = np.mean((data['y'] - y_hat) ** 2)
    mse_estimates.append(mse)
print("MSE", mse_estimates)
```

> **_Answer:_**  The above code does the following steps
> 1.  generates a parabolic synthetic data.
> 2.  Now, a series of model with different polynomial order ranging 1 to 10 is fit on the dataset.
> 3.  For each of the model, the code tries to do a 20-fold CV. But instead of averaging the MSE of all the iteration in k-fold CV, the code tries to take the MSE of the last iteration.
> 4.  Finally, the MSE of each model is added to mse_estimates to see which polynomial degree fits the data the best.

```python
# BEGIN SOLUTION

# Simulate some data
np.random.seed(0)
n = 100
order = 2
x = np.random.uniform(-4, 4, n)
y = np.random.normal(loc=x**order, scale=0.25, size=n)
data = pd.DataFrame({'x': x, 'y': y})
mse_estimates = []


#  Returns the kFold based on no. of fold.
def get_kfold(n_folds):
    """
    Returns the  kFold based on no. of fold given.
    """
    # KFold cross-validation
    kf = KFold(n_splits=n_folds, shuffle=True, random_state=1)
    return (kf)


#  Returns the polynomial fit data of given degree.
def get_polyfit_data(data, degree):
    """
    Returns the polynomial fit data of given degree.
    """
    poly = PolynomialFeatures(degree)
    X_train = poly.fit_transform(np.array(data).reshape(-1,1))
    lr=LinearRegression()
    return (X_train)


# Polynomial degrees 
poly_degrees = range(1, 11) 

# Loop over polynomial degrees
for q in poly_degrees:

    #Get Polynomial fit model of the degree
    X_train = get_polyfit_data(data['x'], q)
    lr=LinearRegression()
    
    # k-fold cross validation on the model
    cve=cross_validate(lr,X_train, data['y'],scoring='neg_mean_squared_error',cv=get_kfold(20),return_train_score=True)

    # Average the k-iteration MSE for each polynomial degree model
    mse = np.mean(np.absolute(cve['test_score']))
    mse_estimates.append(mse)
    
    
print("MSE for each polynomial degree\n", mse_estimates)
# END SOLUTION
```

```python
#Plot MSE vs polynomial Degree model

fig,ax=plt.subplots(figsize=(6,6))
ax.plot(poly_degrees,mse_estimates)
ax.set_xlabel('Degree',fontsize=20)
ax.set_ylabel('MSE',fontsize=20)
ax.set_title('MSE VS Degree',fontsize=25)
```

> As we can see in the above plot, MSE drops when changing the degree from 1 to 2. For degree > 2, there is not much difference which can be seen in MSE. So, we can build the model with polynomial degree of 2 

```python
# Visualising the Polynomial Regression results
# Selected model of degree 2
X_train = get_polyfit_data(data['x'], degree=2)

# OLS polynomial degree 2 model fitting the dataset
ols = sm.OLS(data['y'], X_train).fit()

#Gathering the predictions and actual data in a single dataframe
result = pd.DataFrame({'x':data['x'],'y':data['y'],'y-pred': ols.fittedvalues })
result = result.sort_values('x')

# Plot the data points 
plt.scatter(result.x, result.y, color='blue')

# plotting fitted model
plt.plot(result.x, result['y-pred'], color='red')
plt.title('Polynomial Regression')
plt.xlabel('x')
plt.ylabel('y')
 
plt.show()
```

Bootstrap the following dataset (`n = 1000`) to obtain median and 95% CI for parameter estimates (slope and intercept) summarizing the relationship between `x` and `y_measured.` What happens with median parameter estimates when you examine `y` instead? Provide some visualizations to explain your results.

```python
np.random.seed(1)
nobs = 1000
x = np.random.normal(size=nobs)
y = 5 - 2.25 * x
y_measured = y + np.random.normal(size=nobs)

# BEGIN SOLUTION

#  Returns the 95% confidence interval 
def conf_int(data):
    """
    Returns the 95% Confidence Interval of the data.
    """
    CI=(np.percentile(data,2.5),np.percentile(data,97.25))
    return (CI)


# Define a function bootOLS to return the bootstrap parameter coefficients 
def bootOLS(df, num_samples=1000):
    """ 
    Returns the alpha statistic for num_saples of elements from dataframe df.
    """
    # make a num_samples random choice of indices WITH REPLACEMENT
    indices = np.random.choice(df.index, num_samples, replace=True)
    X = sm.add_constant(df.X[indices])
    y = df.Y[indices]
    
    results = sm.OLS(y,X).fit()
    
    return results.params


###################### Bootstrap OLS with x and y_measured #########################
# call the bootOLS function to get the parameter estimates of the bootstrapped data

#Create Dataframe with x and y
df = pd.DataFrame({'X':x,'Y':y_measured})

beta_0 = []
beta_1 = []
for _ in range(1000):
    params = bootOLS(df)
    beta_0.append(params.iloc[0])
    beta_1.append(params.iloc[1])


#95% Confidence Interval of Beto_0 and Beta_1
mean_beta_0=np.mean(beta_0)
mean_beta_1=np.mean(beta_1)
median_beta_0=np.median(beta_0)
median_beta_1=np.median(beta_1)
ci_beta_0 = conf_int(beta_0)
ci_beta_1 = conf_int(beta_1)

print("--------Parameter estimates when examined with y_measured------------")
print("Median of beta_0 is ",median_beta_0)
print("Standard deviation of beta_0 is ",np.std(beta_0))
print("Confidence Interval of beta_0 is ",ci_beta_0)
print("\n \nMedian of beta_1 is ",median_beta_1)
print("Confidence Interval of beta_1 is ",ci_beta_1)
print("Standard deviation of beta_1 is ",np.std(beta_1))

###################### Bootstrap OLS with x and y#########################

df = pd.DataFrame({'X':x,'Y':y})

y_beta_0 = []
y_beta_1 = []
for _ in range(1000):
    params = bootOLS(df)
    y_beta_0.append(params.iloc[0])
    y_beta_1.append(params.iloc[1])


#95% Confidence Interval of  Beto_0 and Beta_1
y_mean_beta_0=np.mean(y_beta_0)
y_mean_beta_1=np.mean(y_beta_1)
y_median_beta_0=np.median(y_beta_0)
y_median_beta_1=np.median(y_beta_1)
y_ci_beta_0 = conf_int(y_beta_0)
y_ci_beta_1 = conf_int(y_beta_1)

print("\n \n--------Parameter estimates when examined with y------------")
print("Median of beta_0 is ",y_median_beta_0)
print("Confidence Interval of beta_0 is ",y_ci_beta_0)
print("\n \nMedian of beta_1 is ",y_median_beta_1)
print("Confidence Interval of beta_1 is ",y_ci_beta_1)

# END SOLUTION
```
