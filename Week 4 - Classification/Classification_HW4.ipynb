  b) Use the full data set to perform a logistic regression with `Direction` as the response and the five `lag` variables plus `Volume` as predictors. Print the regression results. Do any of the predictors appear to be statistically significant? If so, which ones? 

```python
# BEGIN SOLUTION

# Encoding Direction as 0 and 1 
y =  np.array([1 if el=='Up' else 0 for el in week_df.Direction.values])


# Fitting Logistic Regression Model
glm_weekly = sm.GLM.from_formula('y ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 +  Volume ', 
                                 data = week_df,
                                 family=sm.families.Binomial()).fit()
glm_weekly.summary()
# END SOLUTION
```

> **Answer:** Based on the p-value, Only Lag 2 variable is significant which has p-value of 0.118 less than 0.05.

c) Compute the __confusion matrix__ and overall fraction of correct predictions. Explain what the confusion matrix is telling you about the types of mistakes made by logistic regression.

```python
# BEGIN SOLUTION
# Predicting the target using the model
y_pred_prob = glm_weekly.predict()
y_pred = np.array(["Up" if el > 0.5 else "Down" for el in y_pred_prob])

# confusion Matrix
cm = metrics.confusion_matrix(week_df['Direction'], y_pred)
# END SOLUTION

# Plot the confusion Matrix
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ["Down","Up"])
cm_display.plot()
plt.show()

print("Overall fraction of correct predictions", (cm[0,0] + cm[1,1])/(np.sum(np.array(cm))))
```

> **Answer:**
> The model performs well in predicting the market up movement, which can be interpreted using **Recall**(Out of all actual positive target class, how many were predicted correctly) - TP/(TP+FN) = 557/557+48 = 0.92. So, 92% of the time, the model predicts market up movement correctly.
>
> The model performs poorly in predicting market down movement, which can be interpreted using the parameter **Specificity** (Out of all actual negative target class, how many were predicted correctly) - TN /(TN + FP) = 54/54+430 = 0.11. Only 11% of the market down predictions were made correctly.
>
> There are two type error rate.
> 
> Type 1 error is when the model predicts a actual negative(market down) class to be a positive (market up) target class. FP/ (TN+FP) = 430/54+430 = .88. Around 80% of the actual (negative) market down values were predicted incorrectly to be market positive (positive) target class
>
> Type II error is when the model predicts a actual positive (market up) class to be a negative(market down) target class. FN/ (FN + TP) = 48/48+557 = 0.07. Only 7% of the actual  market up (positive) values were predicted incorrectly to be (negative) market down target class
>
> So, the model is good in predicting the market up movement with a _Recall_ of 92% and _Type II Error_ of 7%.  

  d) Now fit the logistic regression model using a training data period from __1990 to 2008__, with `Lag2` as the only predictor. Compute the confusion matrix and the overall fraction of correct predictions for the held out data (that is, the data from __2009 and 2010__).

```python
# BEGIN SOLUTION

# Filter Dataset with data from 1990 to 2008 into training
train_df = week_df[(week_df['Year'] >= 1990) & (week_df['Year'] <= 2008)]
print("Number of records within training data period from 1990 to 2008 is", len(train_df))

y_train =  np.array([1 if el=='Up' else 0 for el in train_df.Direction.values])

# Filter Dataset with data from 2009 and 2010 into testing
test_df = week_df[(week_df['Year'] > 2008)]
X_test = sm.add_constant(test_df['Lag2'])
y_test = np.array([1 if el=='Up' else 0 for el in test_df.Direction.values])

# Training the model using Logistic Regression
train_df_model = sm.GLM.from_formula('y_train ~ Lag2  ', 
                                 data = train_df,
                                 family=sm.families.Binomial()).fit()

print("Model summary on filtered Dataframe with only `Lag2` as predictor \n", train_df_model.summary())

# Predictions on Test dataset
y_pred_prob = train_df_model.predict(X_test)
y_pred = np.array(["Up" if el > 0.5 else "Down" for el in y_pred_prob])

# confusion Matrix
cm = metrics.confusion_matrix(test_df['Direction'], y_pred)
print("confusion matrix \n", cm)
# END SOLUTION

# Plot the confusion Matrix
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = cm, display_labels = ["Down","Up"])
cm_display.plot()
plt.show()

print("**Overall fraction of correct predictions ", (cm[0,0] + cm[1,1])/(np.sum(np.array(cm))), " **")

# END SOLUTION
```

$f_{yes}(x) = \frac{e^{-(x - \mu_{yes})^{2}/(2* 36) }} {6\sqrt{2\pi}} = \frac{e^{-(x - 10)^{2}/(72) }} {6\sqrt{2\pi}}  $
