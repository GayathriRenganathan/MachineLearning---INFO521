Suppose that a curve $\hat{g}$ is computed to smoothly fit a set of `n` points using the following formula:

  $$
  \hat{g} = \arg\min_{g}(\sum_{i=1}^n (y_i - g(x_i))^2 +  \lambda \int [g^{(m)} (x)]^2 dx )
  $$
where $g^{(m)}$  represents the $m^{th}$ derivative of $g$ (and $g^{(0)} = g$). Provide a description (or sketch) of $\hat{g}$ in each of the following scenarios. **Again, don't forget that $m^{th}$ is related to the derivative of $g$**.

  (a) $\lambda$ = $\infty$, $m = 0$.

   > **_Answer:_**  When m = 0, if $g^{(0)} (x) = g (x)$
> 
>When  $\lambda$ = $\infty$, the penalty term $\lambda \int [g^{(0)} (x)]^2 dx$ becomes more important and the loss term is ignored, So, the function is minimized when the penalty term is zero.
> This forces $ g^{(0)} (x)  -> 0$, so  g(x) = 0.
> 
> So, $ \hat{g} = 0$

```python
# Example sketch 
# Plot of g^ = 0 
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,10,1)
y = np.full(10,0)
plt.title("Plot of $ \\hat{g}$")
plt.plot(x,y,'-r');
```

> This forces $ g^{(1)} (x)  -> 0$, so the g(x) = c, as the first derivative of the constant is zero
> 
> So, $ \hat{g} = c$

> **_Answer:_**
> 1) Step Function - It is a function that can be used in the dataset, where the data is constant for a specfic range and then followed by a sudden spike resembling a step. Step function is likely the preferred option when the data in the predictor has a clear and discrete cuts rather than a smooth/continuous change.
>> For example, we can consider telecom data where we want to analyze how the plan cost with the usage.
>> - If the usage is less than 1hr, then the plan cost is \$10.
>> - If the usage is from 1hr to 2hr, then the plan cost is \$25
>> - If the usage is more than 3hr, then the plan cost is 50$ .
> The cost of the plan shows a sudden spike at different interval, in such cases step function can be used.
> 2) local regression - Local regression is useful in understanding the complex and non-linear relationship between the predictor and the response. The relationship between the predictor and the response that are difficult to be captured by a global functional form and has varying local functional forms can be modeled using local regression. In other words, Local regression is useful when the impact of the predictor over the response varies abruptly and it is difficult to be captured by a single global function.
>>   For example, if we want to predict the sales of puffer jacket over the entire year. The sales of the jacket would be very high during winter
>period, and the sales of the jacket would be comparitively less during fall and the sales would be nearly zero during summer. Such varying relationships are difficult to be captured by a single linear regression model, we need to fit various local models to understand the relationship.

> 3) GAMs - It is called Generalized Additive Model. It is used when we want a model to fit the complex, non-linear relationship in the dataset and also interpretability of the model is also important. GAM does this by adding the non-linear fitted model for each predictor with respect to the response. GAM is primarily useful when each predictor exhibits a complex non-linear relationship with the response.
>>   For example, if we want to predict the health outcome based on age and physical activity. If each of the predictor(age and physical activity) has a non-linear relationship with the health outcome, in such cases GAM can be used to model each of the predictor seperately and them adding them up for the combined effect and hence the interpretability of the model is also maintained.

a) Split the data into a training set and a test set (50:50). Using out-of-state tuition as the response and the other variables as the predictors, perform *forward* stepwise selection on the training set in order to identify a satisfactory model that uses just a subset of the predictors.

```python
# BEGIN SOLUTION

# Reading the College dataset
college_df = pd.read_csv("College.csv", index_col = 0)
college_df.head()

# END SOLUTION
```

```python
# Convert categorical value using one-hot encoding
college_df = pd.get_dummies(data = college_df, columns=['Private'], drop_first=True, dtype = int)
college_df.head()
```

```python

#Split the dataset into train and test
X = college_df.drop(['Outstate'], axis=1)
Y = college_df['Outstate']
xtrain, xtest, ytrain, ytest = train_test_split(X,Y,test_size=0.5, random_state=1)
```

```python
#list of models with increasing number of features upto max_features
def forward_step_select(df, predictors, response, max_features=17):

    # Return the OLS fit model and RSS for the given features
    def linear_model_fit(features):
        # Create dataset for the model
        X = sm.add_constant(df[features])
        #Fitting the OLS model
        model = sm.OLS(response,X).fit()
        RSS = model.ssr
        return (model, RSS)
    
    # Computes the RSS of possible new models and returns the model with the lowest RSS.
    def update_model(best_features, remaining_features):
        results = []
        
        for feature in remaining_features:
            results.append(linear_model_fit(best_features + [feature]))
            
        # select model with the lowest RSS
        new_model = sorted(results, key= itemgetter(1)).pop(0)[0]
        new_features = list(new_model.params.index)[1:]
        
        return new_features, new_model
    
    # Create list to hold models, model features and the remaining features to test
    models = []
    best_features = []
    remaining_features = predictors
    
    while remaining_features and len(best_features) < max_features:
        
        # get the best new feature set from update_model
        new_features, new_model = update_model(best_features, remaining_features)
        # update the best features to include the one we just found
        best_features = new_features  
        # reduce the available features for the next round
        remaining_features =  [feature for feature in predictors if feature not in best_features]
        
        # append the new_features and model so we can compare models with different features later
        models.append((new_features,new_model))
        
    return models
```

```python
# Forward Stepwise Selection
models = forward_step_select(df=xtrain, predictors=list(xtrain.columns), response= ytrain,max_features=17 )
```

```python
# Three metrics to evaluate the best model AIC, BIC and adj, R^2
aics = [models[x][1].aic for x in range(len(models))]
bics = [models[x][1].bic for x in range(len(models))]
r_adj = [models[x][1].rsquared_adj for x in range(len(models))]

# find the mins/maxes
min_aic_index, min_aic = min(enumerate(aics), key=itemgetter(1))
min_bic_index, min_bic = min(enumerate(bics), key=itemgetter(1))
max_radj_index, max_radj = max(enumerate(r_adj), key=itemgetter(1))

num_predictors = np.linspace(1,len(models),len(models))
# Create a plot
fig,(ax1, ax2) = plt.subplots(1,2,figsize=(16,4))
# Add test MSE estimates
ax1.plot(num_predictors, aics, 'r', marker='o', label='AIC');
ax1.plot(num_predictors, bics, 'b', marker='o', label='BIC')

# add the minimums to the axis
ax1.plot(min_aic_index+1, min_aic, 'gx', markersize=20, markeredgewidth=1)
ax1.plot(min_bic_index+1, min_bic, 'gx', markersize=20, markeredgewidth=1)

# Labels and Legend
ax1.set_xlabel('Number of Predictors');
ax1.set_ylabel('Test MSE');
ax1.legend(loc='best');

# Add Adj R**2
ax2.plot(num_predictors, r_adj,'k', marker='o')
ax2.plot(max_radj_index+1, max_radj, 'gx', markersize=20, markeredgewidth=1)
ax2.set_xlabel('Number of Predictors');
ax2.set_ylabel('Adjusted R**2');
```

> 9 Predictor models seems to be the best, after that the $R^2 $ value seems to be almost constant and the AIC values are increasing after 9 predictors.

```python
# 9 feature model
best_predictors = list(models[8][1].params.index[1:])
print("Best Predictors", best_predictors)
print("Best Predictors model summary", models[8][1].summary())
y_pred =  models[8][1].predict(sm.add_constant(xtest[best_predictors]))
print('******Forward Subset Model MSE: %.2f ***********' %mean_squared_error(ytest, y_pred)) 
```
