#### Question 3

a) Discuss the implications that uncertainty in $X$ would have on parameter estimates based on linear regression models. Should we care about error in $X$ when fitting OLS regressions? 

> **_Answer:_**  Uncertainity in X in linear regression model leads to uncertainity in predictor estimates / co-efficients. If the uncertainity is more in X, then we cannot rely on the model estimates, as the model will not reflect the true relationship between the predictor and the target variable.
>
> This will lead to wider confidence interval , incorrect predictor estimates and an unreliable model
>
> As the uncertainity in X can cause serious concerns, we should try to fix this by correcting the uncertainity in X, by using the average of the predictor variable recorded over time rather than using single predictor measurement value or by using an model which can handle the uncertainity in X.

b) Provide a brief discussion on the concept of regression dilution.

> **_Answer:_**  **Regression Dilution** is the statistical phenomenon, caused by the measurement error / uncertainity in Predictor, biasing the linear regression slope towards zero. This is also called Regression attenuation.
> This will result in underestimation of the relationship associated with the predictor and the response.
>
> Such scenario can occur when there is a measurement error in the predictor variable. This could be avoided by taking the average value of the measurement taken over time, instead of single measurement of the predictor variable.

c) Are there any linear regression models that account for uncertainty in $X$ and $Y$? Please do some independent research and cite your sources.

> **_Answer:_**  Models like **Error in Variables (EIV), Bayesian Regressor, Least Square** models, tries to handle the uncertainity in X and Y in the linear regression, thereby we can rely on the predictor estimates and the model will be reliable providing the true association between the predictor and target variable. 
> 
> EIV models handles the uncertainity using techniques like Instrumental Variables , Maximum likelihood. EIV models expects an additional variable (instrument), which is highly correlated with the predictor X and not with the error in X [arXiv:1906.03989v1 [cs.LG] 10 Jun 2019 - https://arxiv.org/pdf/1906.03989] .
>
> Maximum Likelihood Estimation is a statistical technique  to find parameter estimates that maximize the likelihood of observing the data given the assumed error structure. 

> Sources:
> 1. arXiv:1906.03989v1 [cs.LG] 10 Jun 2019 - https://arxiv.org/pdf/1906.03989
> 2. arXiv:2108.02140v1 [math.ST] 4 Aug 2021 - https://arxiv.org/pdf/2108.02140
 

d) Distinguish between least squares and maximum likelihood in the context of linear regression models.

> **_Answer:_**  **Least Squares** method tries to identify the model parameters that best fit the dataset, thereby **minimizing the error in prediction**.
> 1. The goal in LS method is to **minimize the Sum of squared errors(RSS)** of the model predictions.
> 2. Least Square method is computationally costly in the presence of large datasets.
> 3. Least Square method uses the _derivative of the error_ so as to minimize it.
> 4. So, Least Square Method works on basis of Calculus

> The **Maximum Likelihood** method tries to find the parameters of the model, so as to **maximize the likehood of observing the sample**.
> 1. The goal is to **maximize the probability of observing your dataset**, given the parameters of that model.
> 2. Maximum Likelihood uses _log_ function to maximize the likelihood function.
> 3. Maximum Likelihood works on the basis of probability

> When the residuals are normally distributed, both Least Square and Maximum likelihood leads to same parameter estimates, meaning minimizing the quadratic distance in OLS is the same as maximizing the log-likelihood.

c) Produce diagnostic plots of the least squares regression fit using any suitable package or function. Comment on any issues you observe with the fit.

```python
# BEGIN SOLUTION
# model values
model_fitted_y = model.fittedvalues
# model residuals
model_residuals = model.resid
# normalized residuals
model_norm_residuals = model.get_influence().resid_studentized_internal
# absolute squared normalized residuals
model_norm_residuals_abs_sqrt = np.sqrt(np.abs(model_norm_residuals))
# absolute residuals
model_abs_resid = np.abs(model_residuals)
# leverage, from statsmodels internals
model_leverage = model.get_influence().hat_matrix_diag
# cook's distance, from statsmodels internals
model_cooks = model.get_influence().cooks_distance[0]

#Residuals vs Fitted plot
plot_lm_1 = plt.figure()
plot_lm_1.axes[0] = sns.residplot(x=model_fitted_y, y=auto.columns[0], data=auto,
                          lowess=True,
                          scatter_kws={'alpha': 0.5},
                          line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8})

plot_lm_1.axes[0].set_title('Residuals vs Fitted')
plot_lm_1.axes[0].set_xlabel('Fitted values')
plot_lm_1.axes[0].set_ylabel('Residuals');
# END SOLUTION
```

```python
# Plotting Normal Q-Q
QQ = ProbPlot(model_norm_residuals)
plot_lm_2 = QQ.qqplot(line='45', alpha=0.5,  lw=1)
plot_lm_2.axes[0].set_title('Normal Q-Q')
plot_lm_2.axes[0].set_xlabel('Theoretical Quantiles')
plot_lm_2.axes[0].set_ylabel('Standardized Residuals');
# annotations
abs_norm_resid = np.flip(np.argsort(np.abs(model_norm_residuals)), 0)
abs_norm_resid_top_3 = abs_norm_resid[:3]
for r, i in enumerate(abs_norm_resid_top_3):
    plot_lm_2.axes[0].annotate(i,
                               xy=(np.flip(QQ.theoretical_quantiles, 0)[r],
                                   model_norm_residuals[i]));
```

```python
#Scale-Location plot
plot_lm_3 = plt.figure()
plt.scatter(model_fitted_y, model_norm_residuals_abs_sqrt, alpha=0.5);
sns.regplot(x=model_fitted_y,y= model_norm_residuals_abs_sqrt,
              scatter=False,
              ci=False,
              lowess=True,
              line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8});
plot_lm_3.axes[0].set_title('Scale-Location')
plot_lm_3.axes[0].set_xlabel('Fitted values')
plot_lm_3.axes[0].set_ylabel('$ sqrt of  {|Standardized Residuals|}$');

```

```python
#Residuals vs Leverage plot
plot_lm_4 = plt.figure();
plt.scatter(model_leverage, model_norm_residuals, alpha=0.5);
sns.regplot(x=model_leverage,y= model_norm_residuals,
              scatter=False,
              ci=False,
              lowess=True,
              line_kws={'color': 'red', 'lw': 1, 'alpha': 0.8});
plot_lm_4.axes[0].set_xlim(0, max(model_leverage)+0.01)
plot_lm_4.axes[0].set_ylim(-3, 5)
plot_lm_4.axes[0].set_title('Residuals vs Leverage')
plot_lm_4.axes[0].set_xlabel('Leverage')
plot_lm_4.axes[0].set_ylabel('Standardized Residuals');


# annotations
leverage_top_3 = np.flip(np.argsort(model_cooks), 0)[:3]
for i in leverage_top_3:
  plot_lm_4.axes[0].annotate(i,
                             xy=(model_leverage[i],
                                 model_norm_residuals[i]));
```

> An ideal Residuals vs Fitted plot looks like a random noise, indicating no patterns in the scatterplot. And, the red line should be horizonal.
>
> The bow shaped red line indicates that model underfits the data, failing the capture the non-linear patterns in the data
>
>  **Normal Q-Q** plot shows if the residuals are normally distributed. Ideally, all the points should be aligned well on the straight line. The plot shows that few points that deviate from the red line, indicating that few residuals points are not normally distributed.
>
>  **Scale - location** plot shows the spread of the residuals and helps to find if the residuals shows a non-constant variance(heteroscedasticity). An ideal heteroscedastic plot has a sideways “V” shape. Our graph shows the spread is more on the right vs left and middle. This might be caused by not capturing the non-linearities in the model.

Consider the fitted values that result from performing linear regression without an intercept. In this setting, the ith fitted value takes the form 
$$
\hat{y}_i = x_i \hat{\beta}
$$

Where, 
$$
\hat{\beta} = \left( \sum_{i=1}^n x_{i} y_{i} \right) / \left( \sum_{i'=1}^n x_{i'}^2 \right)
$$

Show that we can write
$$
\hat{y}_i = \sum_{i'=1}^n a'_i y'_i
$$

What is $a'_i$?

> **_Answer:_**  

Substitute value of $\hat{\beta}$ in    $\hat{y}_i = x_i \hat{\beta}$
$$ \hat{y}_i = \left ( x_{i} \sum_{i'=1}^n x_{i'} y_{i'} \right) / \left( \sum_{i''=1}^n x_{i''}^2 \right) $$
$$ \hat{y}_i = \left (  \sum_{i'=1}^n x_{i} x_{i'} y_{i'} \right) / \left( \sum_{i''=1}^n x_{i''}^2 \right) $$
This takes the form, 
$$
\hat{y}_i = \sum_{i'=1}^n a'_i y'_i
$$

Hence,  $$a'_i = \left (   x_{i} x_{i'} \right) / \left( \sum_{i''=1}^n x_{i''}^2 \right) $$
